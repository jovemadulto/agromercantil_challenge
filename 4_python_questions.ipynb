# Questões Python
- Visualização em Streamlit
    - Crie uma aplicação em Streamlit que:
        - Exiba dashboards com preços e tendências
        - Permita filtros por produto, região e período
        - Mostre gráficos interativos

Executável a partir do comando:

`python streamlit run streamlit_dashboards.py`

## Visualizações

Foram escolhidas as seguintes visualizações:
- Gráfico de linha
    - Para demonstrar a evolução do preço nas séries temporais
- Gráfico de caixa
    - Para demonstrar a distribuição do preço por produto e identificar outliers
- Gráfico de barras
    - Para representar o histograma de preços observados durante o período
- Tabelas
    - Ajudam a visualização das anomalias, como preços negativos e outliers
